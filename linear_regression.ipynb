```python
import torch
from tqdm import tqdm
```

```python
N = 10 # datapoints

D_in = 1
D_out = 1

X = torch.rand(N, D_in)

true_W = torch.tensor([[2.0]])
true_b = torch.tensor([[1.0]])
y_true = X @ true_W + true_b + torch.rand(N, D_out) * 0.1 # little noise

print(f"true_W: {true_W}")
print(f"true_b: {true_b}")
print(f"y_true: {y_true}")
print(f"X: {X}")
```

```python
lr = 0.01
epochs = 10000

W, b = torch.randn(1, 1, requires_grad=True), torch.randn(1, requires_grad=True)
print(f"W: {W}")
print(f"b: {b}")
```

```python
for epoch in tqdm(range(epochs)):
    if W.grad is not None:
        W.grad.zero_()
    if b.grad is not None:
        b.grad.zero_()

    # Forward pass
    y_hat = X @ W + b

    # Loss calculation
    loss = torch.mean((y_hat - y_true) ** 2)

    # Backward pass
    loss.backward()

    # Update parameters (in-place to preserve requires_grad)
    with torch.no_grad():
        W -= lr * W.grad
        b -= lr * b.grad

    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Loss={loss.item()}, W={W.item()}, b={b.item()}")

print(f"Final Parameters: W={W.item()}, b={b.item()}")
```
